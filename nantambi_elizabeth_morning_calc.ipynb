```python
#import liabraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#control Display of seaborn data
pd.options.display.max_columns=50
sns.set(style="darkgrid", rc={"axes.facecolor":"#CAF1DE"})
```

```python
#loading  dataset
df2 = pd.read_csv('bse_dataset2.csv')
```

```python
#undestanding of data structures
#checking shapes of data structures- understanding row and columns
df2.shape
```

```python
#check null values
df2.isnull().sum().sort_values(ascending=False)
```

```python
#check null values in percentage
for  i in df2.columns:
    if df2[i].isnull().any():
        print(i, "---",df2[i].isnull().sum()*100/df2.shape[0],"%")
```

```python
#checking null values in the figure
import missingno as msno
msno.bar(df2,color="red",fontsize=24)
```

```python
#check for unqiue values
for i in df2.columns:
    print(i,"---",df2[i].unique(),"---",df2[i].unique())
```
